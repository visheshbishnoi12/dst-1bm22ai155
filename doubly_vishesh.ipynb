```python
class Node:
    def __init__(self, data):
        self.data = data
        self.nref = None
        self.pref = None

class doublyLinkedList:
    def __init__(self):
        self.head = None

    def dis(self):
        if self.head is None:
            print("LinkedList is empty")
        else:
            n = self.head
            while n is not None:
                print(n.data, "-->", end=" ")
                n = n.nref

    def rev(self):
        print()
        if self.head is None:
            print("Linked is empty")
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            while n is not None:
                print(n.data, "-->", end=" ")
                n = n.pref

    def add_empty(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("Linked List is not empty")

    def addb(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.nref = self.head
            self.head.pref = new_node
            self.head = new_node

    def adde(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            n.nref = new_node
            new_node.pref = n

    def addaf(self, existing_value, data):
        new_node = Node(data)
        if self.head is None:
            print("Linked List is empty")
        else:
            n = self.head
            while n is not None:
                if n.data == existing_value:
                    new_node.nref = n.nref
                    if n.nref is not None:
                        n.nref.pref = new_node
                    n.nref = new_node
                    new_node.pref = n
                    if new_node.nref is not None:
                        new_node.nref.pref = new_node  # Update the pref reference of the node after the new node
                    break
                n = n.nref

    def delb(self):
        if self.head is None:
            print("Linked List is empty, cannot delete from the beginning")
        else:
            self.head = self.head.nref
            if self.head is not None:
                self.head.pref = None

    def delv(self, value):
        if self.head is None:
            print("Linked List is empty, cannot delete by value")
        else:
            n = self.head
            while n is not None:
                if n.data == value:
                    if n.pref is not None:
                        n.pref.nref = n.nref
                    if n.nref is not None:
                        n.nref.pref = n.pref
                    break
                n = n.nref

    def dele(self):
        if self.head is None:
            print("Linked List is empty, cannot delete from the end")
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            if n.pref is not None:
                n.pref.nref = None
            else:
                self.head = None

dll = doublyLinkedList()

dll.addb(30)
dll.addb(20)
dll.addb(10)

dll.dis()

dll.rev()
print()
dll.adde(40)
dll.dis()
print()
dll.addaf(20, 50)
print()
dll.dis()
dll.delb()
print()
dll.dis()

dll.delv(40)
print()
dll.dis()

dll.dele()
dll.dis()
```
